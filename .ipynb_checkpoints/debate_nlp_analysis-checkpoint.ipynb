```python
with open("./data/presidentialdebate09292020.txt") as f:
    raw_text = f.read()
```

```python
all_speaker_texts = raw_text.split("\n") 
```

```python
speaker_texts_data = []

all_speaker_texts = [_text for _text in all_speaker_texts if _text != ""]

n = len(all_speaker_texts)

for i, speaker_text in enumerate(all_speaker_texts):
    if "Chris Wallace:" in speaker_text and i < n:
        text = all_speaker_texts[i+1]
        unique_words = list(set(text.split()))
        temp_text = {
            "index": i, 
            "speaker": "Chris Wallace", 
            "text": text,
            "words": unique_words,
            "word_count": len(unique_words),
            "x": i,
            "y": 0, 
            "color": "yellow"
        }
        speaker_texts_data.append(temp_text)
    if "President Donald J. Trump:" in speaker_text and i < n:
        text = all_speaker_texts[i+1]
        unique_words = list(set(text.split()))
        temp_text = {
            "index": i, 
            "speaker": "Donald Trump", 
            "text": text,
            "words": unique_words,
            "word_count": len(unique_words), 
            "x": i,
            "y": len(unique_words),
            "color": "red"
        }
        speaker_texts_data.append(temp_text)
    if "Vice President Joe Biden:" in speaker_text and i < n:
        text = all_speaker_texts[i+1]
        unique_words = list(set(text.split()))
        temp_text = {
            "index": i, 
            "speaker": "Joe Biden",
            "text": text,
            "words": unique_words,
            "word_count": len(unique_words),
            "x": i,
            "y": -len(unique_words), 
            "color": "blue"
        }
        speaker_texts_data.append(temp_text)
        
speaker_texts_data = sorted(speaker_texts_data, key=lambda k: k["index"])
```

```python
speaker_texts_data
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import pandas as pd
import json
```

```python
speaker_text_df = pd.read_json(json.dumps(speaker_texts_data))
```

```python
nominee_text_df = speaker_text_df[speaker_text_df["speaker"] != "Chris Wallace"]
```

```python
nominee_text_df
```

```python
plt.figure(figsize=(30,10))
plt.bar(x=range(len(nominee_text_df)), 
        height=nominee_text_df["y"], 
        color=nominee_text_df["color"])

red_patch = mpatches.Patch(color='red', label='Trump')
blue_patch = mpatches.Patch(color='blue', label='Biden')
plt.legend(handles=[red_patch, blue_patch])
```

```python
from wordcloud import WordCloud
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))

STOPWORDS.update(["want", "said", "would", "got", "re", "ve", "don", "look", "going", "think", "know", "people", "well", "way", "year", "say"])
```

```python
trump_text_df = speaker_text_df[speaker_text_df["speaker"] == "Donald Trump"]
trump_all_text = " ".join(trump_text_df["text"].values)
```

```python
# lower max_font_size
wordcloud = WordCloud(max_font_size=1000, stopwords=STOPWORDS).generate(trump_all_text)
plt.figure(figsize=(20, 15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
biden_text_df = speaker_text_df[speaker_text_df["speaker"] == "Joe Biden"]
biden_all_text = " ".join(biden_text_df["text"].values)
```

```python
# lower max_font_size
wordcloud = WordCloud(max_font_size=1000, stopwords=STOPWORDS).generate(biden_all_text)
plt.figure(figsize=(20, 15))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
